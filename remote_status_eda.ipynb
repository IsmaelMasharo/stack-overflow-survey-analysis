```python
dataset = 'survey_report_combined.csv'
df = pd.read_csv(dataset, dtype=object)
df.year = pd.to_datetime(df.year).dt.year
df.original_missing_values_per_row_percentage = df.original_missing_values_per_row_percentage.astype(float)
```

```python
def feature_across_years(feature, year_resume=True, value_counts=True):
    if value_counts:
        df_counts = df[feature].value_counts().to_frame()
        display(df_counts)
    if year_resume:
        feature_df = df.groupby(['year', feature])[feature].count().to_frame()
        feature_df.columns = ['count']
        feature_df['percentage'] = feature_df.groupby(level=0).apply(lambda x: x / float(x.sum())).round(4)
        display(feature_df)        
```
